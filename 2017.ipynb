## Day 10

```python
# dataset = '2017/day10_test.txt'
dataset = '2017/day10.txt'

current_position = 0
skip_size = 0

hash_list = list(range(256))
len_list = len(hash_list)

with open(dataset) as f:
    lengths = [int(i) for i in f.readline().split(',')]

for chunk_size in lengths:
    # select chunk
    work_list = hash_list[current_position:current_position + chunk_size]
    tail = chunk_size - (len_list - current_position)
    if current_position + chunk_size >= len_list:
        tail = chunk_size - (len_list - current_position)
        work_list += hash_list[0:tail]
    # print(work_list)
    # reverse chunk
    work_list.reverse()
    # print(work_list)

    # fill back list with reversed values
    x = chunk_size if current_position + chunk_size < len_list else len_list - current_position
    new_list = hash_list[0:current_position] + work_list[:x]
    if tail > 0:
        new_list = work_list[x:] + new_list[tail:]
    else:
        new_list += hash_list[current_position + chunk_size:]
    hash_list = new_list

    # move variables
    current_position = (current_position + chunk_size + skip_size) % len_list
    skip_size += 1

print(hash_list[:2])
print(hash_list[0] * hash_list[1])  # 23715
```

## Day 11

```python
directions = {
    'n': (0, 1, -1),
    'ne': (1, 0, -1),
    'se': (1, -1, 0),
    's': (0, -1, 1),
    'sw': (-1, 0, 1),
    'nw': (-1, 1, 0),
}
```

```python
# dataset = '2017/day11_test.txt'
dataset = '2017/day11.txt'

with open(dataset) as f:
    lines = f.readlines()
```

### Part 1

```python
# fewest steps to the target
for line in lines:
    final_direction = (0, 0, 0)
    for step in line.strip().split(','):
        d = directions[step]
        final_direction = (final_direction[0] + d[0], final_direction[1] + d[1], final_direction[2] + d[2])
    print(max(abs(final_direction[0]), abs(final_direction[1]), abs(final_direction[2])))  # 698

```

### Part 2

```python
# max distance
for line in lines:
    current_position = (0, 0, 0)
    max_distance = 0
    for step in line.strip().split(','):
        d = directions[step]
        current_position = (current_position[0] + d[0], current_position[1] + d[1], current_position[2] + d[2])
        max_distance = max(max_distance,
                           max(abs(current_position[0]), abs(current_position[1]), abs(current_position[2])))
    print(max_distance)  # 1435

```

## Day 12

```python
# dataset = '2017/day12_test.txt'
dataset = '2017/day12.txt'
graph = {}
with open(dataset) as f:
    for line in f.readlines():
        tmp = line.split(' <-> ')
        key = tmp[0].strip()
        values = tmp[1].strip().split(', ')
        graph[key] = values
```

```python
def dfs(graph, start):
    visited = set()
    stack = [start]
    while stack:
        current = stack.pop()
        if current not in visited:
            visited.add(current)
            stack.extend(reversed(graph[current]))
    return visited
```

```python
print(len(dfs(graph, '0')))  # 283
```

### Part 2


```python
groups = 0
while len(graph) != 0:
    to_visit = list(graph.keys())[0]
    visited = dfs(graph, to_visit)
    for key in visited:
        del graph[key]
    groups += 1
print(groups)  # 195
```
