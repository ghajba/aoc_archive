## Day 10

```python
# dataset = '2017/day10_test.txt'
dataset = '2017/day10.txt'

current_position = 0
skip_size = 0

hash_list = list(range(256))
len_list = len(hash_list)

with open(dataset) as f:
    lengths = [int(i) for i in f.readline().split(',')]

for chunk_size in lengths:
    # select chunk
    work_list = hash_list[current_position:current_position + chunk_size]
    tail = chunk_size - (len_list - current_position)
    if current_position + chunk_size >= len_list:
        tail = chunk_size - (len_list - current_position)
        work_list += hash_list[0:tail]
    # print(work_list)
    # reverse chunk
    work_list.reverse()
    # print(work_list)

    # fill back list with reversed values
    x = chunk_size if current_position + chunk_size < len_list else len_list-current_position
    new_list = hash_list[0:current_position] + work_list[:x]
    if tail > 0:
        new_list = work_list[x:] + new_list[tail:]
    else:
        new_list += hash_list[current_position+chunk_size:]
    hash_list = new_list


    # move variables
    current_position = (current_position + chunk_size + skip_size) % len_list
    skip_size += 1

print(hash_list[:2])
print(hash_list[0] * hash_list[1]) # 23715
```

### Part 2

```python
# dataset = '2017/day10_test.txt'
dataset = '2017/day10.txt'

current_position = 0
skip_size = 0

hash_list = list(range(256))
len_list = len(hash_list)

with open(dataset) as f:
    lengths = [ord(i) for i in f.readline()] + [17,31,73,47,23]

for _ in range(64):
    for chunk_size in lengths:
        # select chunk
        work_list = hash_list[current_position:current_position + chunk_size]
        tail = chunk_size - (len_list - current_position)
        if current_position + chunk_size >= len_list:
            tail = chunk_size - (len_list - current_position)
            work_list += hash_list[0:tail]

        if len(work_list) != chunk_size:
            print('error', len(work_list))
            print(work_list)
        # reverse chunk
        work_list.reverse()
        # print(work_list)

        # fill back list with reversed values
        x = chunk_size if current_position + chunk_size < len_list else len_list - current_position
        new_list = hash_list[0:current_position] + work_list[:x]
        if tail > 0:
            new_list = work_list[x:] + new_list[tail:]
        else:
            new_list += hash_list[current_position + chunk_size:]
        hash_list = new_list

        # move variables
        current_position = (current_position + chunk_size + skip_size) % len_list
        skip_size += 1

n=16
dense = []
for i in range(n):
    res = 0
    for a in hash_list[i*n:(i+1)*n]:
        res ^= a
    dense.append(res)

hexa = [hex(i)[2:] for i in dense]
hexa = [i if len(i) == 2 else '0'+i for i in hexa]
print(''.join(hexa)) # 541dc3180fd4b72881e39cf925a50253
```
